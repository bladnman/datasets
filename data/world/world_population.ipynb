```python
pop_df = pd.read_csv("./datasets/data/world/population_by_country.csv")
codes_df = pd.read_csv("./datasets/data/world/country_code_mapping.csv")

pop2_df = pd.merge(pop_df, codes_df, left_on='Country Code', right_on='Alpha-3 code', how = 'left')
pop2_df['Code'] = pop2_df['Alpha-2 code'].astype(str)
pop2_df['Code3'] = pop2_df['Country Code'].astype(str)
pop2_df['Name'] = pop2_df['Country Name'].astype(str)
pop2_df.drop(['Alpha-3 code', 'Country', 'Country Code', 'Country Name', 'Unnamed: 65', 'Indicator Name', 'Indicator Code', 'Alpha-2 code', 'Numeric'], axis=1, inplace=True)
# remove nulls
pop2_df.dropna(subset=['1960'], inplace=True)
pop2_df.dropna(subset=['2019'], inplace=True)
pop2_df = pop2_df[pop2_df['Code'] != "nan"]

# reorder the columns
pop2_df.insert(0, 'Code3', pop2_df.pop("Code3"))
pop2_df.insert(0, 'Code', pop2_df.pop("Code"))
pop2_df.insert(0, 'Name', pop2_df.pop("Name"))

pop2_df.to_csv("./datasets/data/world/population_by_country.csv", index=False)
```
